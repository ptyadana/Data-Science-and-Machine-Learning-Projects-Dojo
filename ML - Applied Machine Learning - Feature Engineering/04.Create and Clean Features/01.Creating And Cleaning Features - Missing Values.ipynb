## Creating And Cleaning Features: Treat Missing Values In The Data

Three primary methods to treat missing values:
1. Impute to the median or mean of the feature
2. Model that feature to predict the missing values
3. Assign it some default value (`Other` or `-999`)

### Read In Data

```python
# Read in data
import pandas as pd

titanic_df = pd.read_csv('../Data/titanic.csv')
titanic_df.head()
```

```python
# Check where we have missing values
titanic_df.isnull().sum()
```

### Fill Missing For `Age`

```python
# Check if Age appears to be missing at random
# between Age missing group and Non Age missing group
titanic_df.groupby(titanic_df['Age'].isnull()).mean()
```

```python
# Fill in missing values for the Age feature
titanic_df['Age_clean'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())

# recheck the total missing values
titanic_df.isnull().sum()

# now we can see that Orignal Age column has missing values and Newly created 'Age_clean'  column has no more missing values.
```

```python
# Take a look at the data one more time
titanic_df.head(10)
```

### Fill Missing For `Embarked`

```python
# Fill in missing values for the Embarked feature
titanic_df['Embarked_clean'] = titanic_df['Embarked'].fillna('O')

# recheck missing values
titanic_df.isnull().sum()
```

```python
titanic_df.head()
```

```python
# Write out data with missing values cleaned up
titanic_df.to_csv('../Data/titanic_no_missing.csv', index=False)
```
