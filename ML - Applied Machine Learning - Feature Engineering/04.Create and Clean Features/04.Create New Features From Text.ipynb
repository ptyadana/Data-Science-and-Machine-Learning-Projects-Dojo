## Creating And Cleaning Features: Create New Features From Text

### Read In Data

```python
# Read in data
import pandas as pd
titanic_df = pd.read_csv('../Data/titanic_transformed.csv')
titanic_df.head()
```

### Create New Title Feature

```python
# Create a title feature by parsing passenger name
titanic_df['Title'] = titanic_df['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())
titanic_df.head()
```

```python
# Look at survival rate by title
titanic_df.pivot_table('Survived', index=['Title', 'Sex'], aggfunc=['count', 'mean'])
```

```python
# Create new CSV with updated data
titanic_df.to_csv('../Data/titanic_title.csv', index=False)
```
