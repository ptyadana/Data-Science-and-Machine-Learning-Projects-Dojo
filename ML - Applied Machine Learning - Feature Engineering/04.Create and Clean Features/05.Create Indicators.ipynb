## Creating And Cleaning Features: Create Indicators

### Read In Data

```python
# Read in data
import numpy as np
import pandas as pd

titanic_df = pd.read_csv('../Data/titanic_title.csv')
titanic_df.head()
```

### Create Indicator Variable For `Cabin`

```python
# What would an indicator variable for Cabin look like?

# we want to look at how the mean ratio of surival based on Group of Missing Cabin and Non-Missing Cabin
titanic_df.groupby(titanic_df['Cabin'].isnull())['Survived'].mean()
```

As from the ratio above, 2/3 of people who had Cabin survived which is significant.

```python
# Create indicator feature
titanic_df['Cabin_ind'] = np.where(titanic_df['Cabin'].isnull(), 0, 1)

titanic_df.head()
```

Now compared to the new feature `Cabin_ind` with original `Cabin`, model doesn't need to know whether cabin is 'C', etc. It can only focus on binary indicators 0, 1 (missing cabin, cabin present)

----------

```python
# Create new CSV with updated data
titanic_df.to_csv('../Data/titanic_cabin_ind.csv', index=False)
```
