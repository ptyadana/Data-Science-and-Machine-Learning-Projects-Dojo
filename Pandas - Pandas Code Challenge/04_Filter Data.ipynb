## Challenge 4.2) Chain Multiple Conditions to narrow down

+ get customers who don't have partners and born after 1990

```python
df.head(1)
```

```python
df.info()
```

```python
# let's quickly check how many unique marital_status are there
df['Marital_Status'].unique()
```

### First Approach

```python
# born after 1990 and don't have partners (Except Together, Married)
no_partner = (df['Marital_Status'] != 'Together' ) & (df['Marital_Status'] != 'Married')
born_after_1990 = df['Year_Birth'] > 1990

customers = df[no_partner & born_after_1990]
customers
```

### Second Approach, using ~

```python
no_partner = ~df['Marital_Status'].isin(['Married', 'Together'])
born_after_1990 = df['Year_Birth'] > 1990
```

```python
# Second Approach
df.loc[no_partner & born_after_1990]
```

## Challenge 4.3) Bitwise Opearations to filter rows

As the business starts to look for opportunity to selling online, so we want to retrieve customers who Buy Online.
+ New column with "Active" and "Non-Active" Customers.
+ Active customers who meet either of the following criterias.
    + customers who have more than 5 web purchases.
    + cstomers who have made more than 5 "Detal" purchases.


```python
df.head(1)
```

```python
df.columns
```

### Approach 1: using np.where for mutliple filters

```python
# Extract customers with more than 5 Web Purchases , 5 Deals Purchases
cust_web_purchases = df['NumWebPurchases'] > 5
cust_deal_purchases = df['NumDealsPurchases'] > 5
```

```python
df['Status'] = np.where(cust_web_purchases | cust_deal_purchases, 'Active', 'Non-Active')
```

```python
df.head()
```

----

### Approach 2: using .loc method

```python
df.loc[(df['NumWebPurchases'] > 5) | ( df['NumDealsPurchases'] > 5), 'ActiveOnlineBuyer'] = 'Active'
df.loc[(df['NumWebPurchases'] <= 5) & ( df['NumDealsPurchases'] <= 5), 'ActiveOnlineBuyer'] = 'Non-Active'
```
