# Comparison Plots 
- jointplot()
- pairplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/StudentsPerformance.csv')
df.head()
```

# jointplot

```python
sns.jointplot(x='math score', y='reading score', data=df)
```

```python
sns.jointplot(x='math score', y='reading score', data=df, alpha=0.5)
```

## kind='hex'

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='hex')
```

## kind='hist'

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='hist')
```

## kind='kde'

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='kde')
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='kde', shade=True)
```

```python
sns.jointplot(x='math score', y='reading score', data=df, hue='gender', palette='Set2')
```

-------

# Pairplot
- pairplot is `CPU intensive`
- so it is important to take out only necessary columns and do the rendering, if you have large dataset.

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df, hue='gender')
```

## diag_kind ='hist'

```python
sns.pairplot(data=df, hue='gender', diag_kind='hist')
```

## corner=True

```python
sns.pairplot(data=df, hue='gender', corner=True)
```
