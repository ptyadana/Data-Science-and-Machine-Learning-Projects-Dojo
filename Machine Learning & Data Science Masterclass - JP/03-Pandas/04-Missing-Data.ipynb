# There are 3 options to handle the missing data
- Keep Data
- Drop Data
- Fill Data

# Drop Data

```python
# drop every rows which has any missing values
df.dropna()
```

```python
help(df.dropna)
```

```python
df.head()
```

```python
# keep rows which as at least 1 value
df.dropna(thresh=1)
```

```python
# keep rows which has at least 5 values (now Hugh Jackman is gone because it only has 4 values)
df.dropna(thresh=5)
```

```python
df.dropna(axis=1) # drop every columns because every columns has missing values
```

```python
df.head()
```

```python
# it only look at 'last_name' column and drop a row which has missing value in that colum 
df.dropna(subset=['last_name'])
```

```python
df.dropna(subset=['post_movie_score', 'pre_movie_score'])
```

# Filling with Interpolation
- data must be in order before interpolate
- really need to understand whether or not this is a valid choice for your data.

```python
airline_tix = {
    'first': 100,
    'business': np.nan,
    'econmy-plus': 50, 
    'economy': 30,
}
```

```python
ser = pd.Series(airline_tix)
```

```python
ser
```

Assuming that `business` class value is between `first` and `economy-plus`, we can interpolate value. Plus, data are in the correct order.

```python
ser.interpolate()
```
