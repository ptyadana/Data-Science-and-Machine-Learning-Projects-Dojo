# groupby() method

Groupby methods are lazy. So it always wait for aggregrate functions to perform further.

Common Options:

    mean(): Compute mean of groups
    sum(): Compute sum of group values
    size(): Compute group sizes
    count(): Compute count of group
    std(): Standard deviation of groups
    var(): Compute variance of groups
    sem(): Standard error of the mean of groups
    describe(): Generates descriptive statistics
    first(): Compute first of group values
    last(): Compute last of group values
    nth() : Take nth value, or a subset if n is a list
    min(): Compute min of group values
    max(): Compute max of group values

```python
# Creates a groupby object waiting for an aggregate method
df.groupby('model_year')
```

```python
# now 'model_year' became index
df.groupby('model_year').mean()
```

```python
df.groupby('model_year').mean()['mpg']
```

# Cross Section
- it allow you to go deep insides of specific level

```python
# key tied to one value only and tellng to look at which level, in this case 'model_year'
year_cyl.xs(key=70, level='model_year')
```

```python
year_cyl.head()
```

```python
# want to get 4 cylinders for every year
# this tell to look at cylinders level and get values with 4

four_cylinders = year_cyl.xs(key=4, level='cylinders')
four_cylinders
```

# Sorting MultiIndex
- it only make sense to sort by the most outer layer, otherwise the relation will break up.

```python
year_cyl.head()
```

```python
# we want to sort mode_year by decedning order
year_cyl.sort_index(level='model_year', ascending=False)
```

```python
# it only make sense to sort by the most outer layer, otherwise the relation will break up.
# we can see that the relation is broken.
year_cyl.sort_index(level='cylinders', ascending=False)
```

# Advanced: agg() method

the funciton name must matched up with the following list

Common Options:

    mean(): Compute mean of groups
    sum(): Compute sum of group values
    size(): Compute group sizes
    count(): Compute count of group
    std(): Standard deviation of groups
    var(): Compute variance of groups
    sem(): Standard error of the mean of groups
    describe(): Generates descriptive statistics
    first(): Compute first of group values
    last(): Compute last of group values
    nth() : Take nth value, or a subset if n is a list
    min(): Compute min of group values
    max(): Compute max of group values

## agg() on a DataFrame

```python
df.agg(['mean', 'std', 'count'])
```

```python
df.agg(['mean', 'std', 'count'])['mpg']
```

# Specify aggregate methods per column | more elegant way
- for each column, which aggregate methods should be applied => passed them in dictionary

```python
df.agg({'mpg': ['mean', 'std'], 'acceleration':['count', 'sum']})
```

# agg() with groupby()

```python
df.head()
```

```python
df.groupby('model_year').agg({'mpg': ['mean', 'std'], 'acceleration':['count', 'sum']})
```
