## PCA with Scikit-Learn


### Scaling Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

## Scikit-Learn Implementation

```python
from sklearn.decomposition import PCA # note that Decomposing Large Feature Space into Smaller one
```

```python
help(PCA)
```

```python
pca_model = PCA(n_components = 2)
```

```python
# fitting stage: calculate Eigen values and Eigen Vectors
# transofrm stage: does the projection step
principal_components = pca_model.fit_transform(scaled_X) 
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component');
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary = load_breast_cancer()
```

```python
cancer_dictionary.keys()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(principal_components[:,0],principal_components[:,1],c=cancer_dictionary['target'])
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

## Fitted Model Attributes

Let's look at all the attributes of the fitted PCA model.

```python
pca_model.n_components_
```

```python
pca_model.components_
```

In this numpy matrix array, each row represents a principal component, Principal axes in feature space, representing the directions of maximum variance in the data. The components are sorted by explained_variance_.

We can visualize this relationship with a heatmap:

```python
df_comp = pd.DataFrame(pca_model.components_, index=['PC1', 'PC2'], columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(20,3),dpi=150)
sns.heatmap(df_comp, annot=True);
```

```python
pca_model.explained_variance_ratio_
```

We can see that 1st PC can explain 44% of our data and 2nd PC explain 19% of our data.

If we sum two of them, we can explain 63% of the original data.

```python
np.sum(pca_model.explained_variance_ratio_)
```

#### Thought Experiment
As thought experiment, Let's see if we set components to 30 (original number of features) and how much our explained_variance_ratio will be.

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(scaled_X)
```

```python
np.sum(pca_30.explained_variance_ratio_)
```

It make sense to have 1.0 (100%) as we are not reducing any features.

## Elbow Plot
+ We can use this method to see around which number of Principal components should we choose.
#### Visualizing Explained Variance across different number of Components

```python
explained_variance = []

for n in range(1, 30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_variance)
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained");
```

This graph makes sense. As we reduce the number of components,  it starts to get harder to explain our data.
