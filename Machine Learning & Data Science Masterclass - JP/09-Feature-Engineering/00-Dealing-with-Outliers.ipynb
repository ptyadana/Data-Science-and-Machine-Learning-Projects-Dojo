# Dealing with Outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Choose a mean,standard deviation, and number of samples

def create_ages(mu=50,sigma=13,num_samples=100,seed=42):

    # Set a random seed in the same cell as the random call to get the same values as us
    # We set seed to 42 (42 is an arbitrary choice from Hitchhiker's Guide to the Galaxy)
    np.random.seed(seed)

    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
    sample_ages = np.round(sample_ages,decimals=0)
    
    return sample_ages
```

```python
sample = create_ages()
```

```python
sample
```

# Visualize and Describe the Data

```python
sns.displot(data=sample, bins=20);
```

```python
sns.boxplot(data=sample);
```

```python
# as we can see from the above box plot, there are some kind of outliers in the lower range. So we want to grab those values
ser = pd.Series(sample)
ser
```

```python
ser.describe()
```

```python
# find 1.5 IQR limit

IQR = 55.25 - 42.00
lower_limit = 42.00 - (1.5 * IQR)
upper_limit = 55.25 + (1.5 * IQR)

lower_limit, upper_limit
```

#### using numpy percentile method to get the range

```python
# getting 25 and 75 percentile of that dataset
q25, q75 = np.percentile(sample, [25, 75])

q25, q75
```

```python
ser[ser > lower_limit]
```

# Trimming or Fixing Based Off Domain Knowledge

If we know we're dealing with a dataset pertaining to voting age (18 years old in the USA), then it makes sense to either drop anything less than that OR fix values lower than 18 and push them up to 18.

```python
ser[ser > 18]
```

```python
# It dropped one person
len(ser[ser > 18])
```

```python
len(ser)
```

```python
# using function to fix age to 18
def fix_values(age):
    if age < 18:
        return 18
    return age
```

```python
ser[ser.apply(fix_values)]
```

```python
df = pd.read_csv('../Data/Ames_Housing_Data.csv')
```

```python
# check the correlation between the features and Saleprice
df.corr()['SalePrice'].sort_values()
```

We can see from the chart that there are some outliers.
- houses with high quality, but selling at bargin prices.
- houses with median quality, but selling at higher prices.

As we can see from the chart again, there are large Living Area houses but selling at lower prices.

```python
df[(df['SalePrice'] < 200000) & (df['Overall Qual'] > 8)]
```

```python
df[(df['SalePrice'] < 200000) & (df['Gr Liv Area'] > 4000)]
```

As we have defined our outliers as above 3 houses, we will drop them.

```python
drop_index = df[(df['SalePrice'] < 200000) & (df['Gr Liv Area'] > 4000)].index
drop_index
```

```python
df = df.drop(drop_index, axis=0)
```
